Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

```

 Understading and Modifying Datastructure

```python
data=pd.read_csv('lsd_math_score_data.csv')
type(data)
#DataFrame
```

```python
x=data['Avg_Math_Test_Score']
x
```

```python
#ADD NEW COLUMN TO THE DATA FRAME

data['Test_Subject']= 'Jennifer Lopez'  #All rows on this column = Jennifer Lopez

data
```

Data Visualization

```python
time=data[['Time_Delay_in_Minutes']]
LSD=data[['LSD_ppm']]
score=data[['Avg_Math_Test_Score']]
```

```python
plt.title("LSD over time")

plt.xlabel("Time",fontsize=14)
plt.ylabel("LSD concentration",fontsize=14)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.style.use('dark_background') #ggplot, grayscale

plt.plot(time,LSD,color='r',linewidth=3)

plt.show()
```

Training Model

```python
#CRIAR REGRESSÃO objeto de linear reg

regr=LinearRegression()
regr.fit(LSD,score)   #Erro: DataSeries, mas precisa ser DataFrame

#regr.fit -> run regression. 'LSD' independente, 'score' dependente- VALORES QUE TENTAREMOS PREVER
```

Training metrics

```python
regr.coef_[0][0] #Acessar valor do array. Coef é array de array
```

```python
regr.intercept_[0]   #Intercept é um array
```

```python
regr.score(LSD,score)  #LSD concentration explains 87% of Math scores
```

Visualizing Regression and scatter data points

```python
plt.title("Arithmetic vs LSD-25")
plt.xlabel('Tissue LSD ppm',fontsize=14)
plt.ylabel('Perfomance Score',fontsize=14)
plt.ylim(25,85)
plt.xlim(1,6.5)
plt.style.use('dark_background')

plt.plot(LSD,regr.predict(LSD)) #regressão prever o valor de Y de LSD. regr.predict devolve y

plt.scatter(LSD,score,alpha=0.7,s=100)


plt.show()
```
